# 범죄 분류별 합계 후 내림차순

```python
crime_2 = pd.read_csv("C:\\Users\\kim a hyun\\Desktop\\crime_2.csv",encoding='cp949',index_col='범죄 분류')
cols = [
    '서울',
    '부산',
    '대구',
    '인천',
    '광주'
 ] #,ascending=True)

col2 = crime_2.loc[:, cols]
col2
```

```python
col2_group = np.sum(col2.groupby('범죄 분류').sum(), axis=1) # axis=1 은 같은 행별로 계산하는 옵션이다.
```

```python
col2_group
```

```python
col2_group = np.sum(col2.groupby('범죄 분류').sum(), axis=1) # axis=1 은 같은 행별로 계산하는 옵션이다.
```

```python
col2_group = pd.DataFrame(col2_group).reset_index()
```

```python
col2_group.rename(columns = {0:'total'}, inplace = True) # inplace = True: 변경 사항을 바로 적용
```

```python
col2_group.sort_values(by='total', ascending=False, inplace = True)
```

```python
col2_group
```
